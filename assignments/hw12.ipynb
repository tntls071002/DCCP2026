```python
import numpy as np
import pandas as pd

works = pd.DataFrame({
    '작가': ['김유정', '김유정', '김유정',
           '현진건', '현진건',
           '이상', '이상',
           '나도향', '나도향',
           '채만식', '이효석'],
    '작품': ['봄봄', '동백꽃', '만무방',
           '운수 좋은 날', 'B사감과 러브레터',
           '날개', '권태',
           '벙어리 삼룡이', '물레방아',
           '레디메이드 인생', '메밀꽃 필 무렵'],
    '연도': [1935, 1936, 1934,
           1924, 1925,
           1936, 1937,
           1925, 1925,
           1934, np.nan],
    '글자수': [12500, 9800, 13600,
            11200, 7400,
            18300, 14100,
            16800, 14500,
            22000, 11800],
})

authors = pd.DataFrame({
    '작가': ['김유정', '현진건', '이상', '나도향', '채만식', '이효석', '염상섭'],
    '생년': [1908, 1900, 1910, 1902, 1902, 1907, 1897],
    '몰년': [1937, 1943, 1937, 1926, 1950, 1942, 1963],
})
```

**Q1-(a)**

```python
print(works.shape)
```

```python
print(works.dtypes)
```

```python
print(works.describe())
```

연도 열은 NaN(결측치)이 포함되어 있어 정수를 저장할 수 없으므로 float64로 변환되었고, 글자수는 결측치가 없어 int64를 유지한다.

**Q1-(b)**

```python
print(works.isna().sum())
```

연도 열에만 결측치가 1개 존재하며, 다른 열에는 결측치가 없다.

**Q1-(c)**

```python
works2 = works.copy()

works2['연도'] = works2['연도'].fillna(1936).astype(int)

print(works2.dtypes)
print(works2.tail(3))
```

결측 연도를 1936으로 채운 뒤 astype(int)를 사용하여 연도 열을 다시 정수형으로 변환하였다.

**Q2-(a)**

```python
# 1930년 이후 작품
after_1930 = works2[works2['연도'] >= 1930]

print(after_1930)
```

```python
# 작가가 김유정 또는 이상
kim_or_lee = works2[works2['작가'].isin(['김유정', '이상'])]

print(kim_or_lee)
```

불 인덱싱과 isin()을 사용하면 여러 조건을 간단하게 필터링할 수 있다.

**Q2-(b)**

```python
def categorize(n: int) -> str:
    if n < 10000:
        return '짧음'
    elif 10000 <= n <= 15000:
        return '보통'
    else:
        return '긴'

works2['분량'] = works2['글자수'].apply(categorize)

print(works2[['작품', '글자수', '분량']])
```

apply()를 사용해 각 작품의 글자수를 기준으로 분량 범주를 추가하였다.

**Q2-(c)**

```python
sorted_works = works2.sort_values(
    by=['연도', '글자수'],
    ascending=[True, False]
)

print(sorted_works)
```

연도는 오름차순, 같은 연도에서는 글자수가 큰 작품이 먼저 오도록 정렬하였다.

**Q3-(a)**

```python
author_stats = works2.groupby('작가')['글자수'].agg([
    'mean',
    'count',
    'sum'
])

print(author_stats)
```

groupby()와 agg()를 함께 사용하면 여러 통계를 한 번에 계산할 수 있다.

**Q3-(b)**

```python
print(works2['분량'].value_counts())
```

‘보통’ 분량 작품이 가장 많고, ‘짧음’ 작품은 가장 적다.

**Q3-(c)**

```python
count_by_group = works2.groupby(['작가', '분량']).size()

print(count_by_group)
```

가장 많은 작품 수를 가진 조합은 (김유정, 보통)이며, 김유정이 중간 정도 분량의 작품을 주로 집필했음을 시사한다.

**Q4-(a)**

```python
full = pd.merge(
    works2,
    authors,
    on='작가',
    how='left'
)

print(full)
print(full.shape)
```

authors에는 있지만 works2에는 없는 작가인 염상섭은 full에 포함되지 않는다. how='left'는 왼쪽 DataFrame(works2)의 행만 기준으로 유지하기 때문이다.

**Q4-(b)**

```python
full['집필 당시 나이'] = full['연도'] - full['생년']

print(full[['작가', '작품', '연도', '생년', '집필 당시 나이']])
```

벡터화 연산을 사용하여 각 작품이 발표될 당시의 작가 나이를 계산하였다.

**Q4-(c)**

```python
avg_age = full.groupby('작가')['집필 당시 나이'].mean().sort_values()

print(avg_age)
```

가장 어린 나이에 작품을 발표한 작가는 나도향이며 평균 23세에 작품을 발표했다. 가장 늦은 나이에 작품을 발표한 작가는 채만식으로 평균 32세였다.
